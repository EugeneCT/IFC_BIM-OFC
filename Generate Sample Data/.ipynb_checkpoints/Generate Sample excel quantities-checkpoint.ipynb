```python
import pandas as pd
import random


import string
```

```python
MC1={
    'Substructure':['Piling', 'Foundation'],
    'Superstructure':['Frame and Slab']
    }

MC2={
    'Piling':['Permanent Pile'],
    'Foundation':['Raft Footing','Pile Cap','Column Base','Wall Footing','Substructure Wall','Substructure Column','Substructure Beam','Substructure Slab'],
    'Frame and Slab':['Structural Wall','RC Column','RC Beam','RC Slab','RC Staircase','RC Slab','Steel Beam','Steel Column','Steel Staircase']
    }
Item={
    'Permanent Pile':[['Type of Pile','Concrete Grade','Pile Diameter'],"Nos"],
    'Raft Footing':[['Concrete Grade'],"m3"],
    'Pile Cap':[['Concrete Grade'],"m3"],
    'Column Base':[['Concrete Grade'],"m3"],
    'Wall Footing':[['Concrete Grade'],"m3"],
    'Substructure Wall':[['Concrete Grade'],"m3"],
    'Substructure Column':[['Concrete Grade'],"m3"],
    'Substructure Beam':[['Concrete Grade'],"m3"],
    'Substructure Slab':[['Concrete Grade'],"m3"],
    'Structural Wall':[['Type of Construction','Concrete Grade','Type of Wall'],"m3"],
    'RC Column':[['Type of Construction','Concrete Grade'],"m3"],
    'RC Beam':[['Type of Construction','Concrete Grade','Post Tension'],"m3"],
    'RC Slab':[['Type of Construction','Concrete Grade','Post Tension'],"m3"],
    'RC Staircase':[['Type of Construction','Concrete Grade'],"m3"],
    'RC Slab':[['Type of Construction','Concrete Grade'],"m3"],
    'Steel Beam':[['Steel Grade','Steel Thickness'],"kg"],
    'Steel Column':[['Steel Grade','Steel Thickness'],"kg"],
    'Steel Staircase':[['Steel Grade','Steel Thickness'],"kg"]
    
}

Subcat={
    'Type of Pile':['Bored Pile','Spun Pile','Micro Pile'],
    'Pile Diameter':['200 mm','350mm','400'],
    'Concrete Grade':['G30','G40','35'],
    'Type of Construction':["Precast","Cast in-situ","Castinsitu"],
    'Type of Wall':["Lift Core Wall","Parapet Wall","Ramp Wall","Planter Wall"],
    'Post Tension':["Yes,1:35","80%","0.8","No",""],
    'Steel Grade':["S355","EN 10025 S275","BS4360 Grade 43A"],
    'Steel Thickness':["152 x 152 x 23 UC","5"]
}



```

```python
columns=["Cost Code","Main Category 1","Main Category 2","Item","Sub Category 1","Sub Category 2","Sub Category 3","Unit","Rate","Description"]
# df=pd.DataFrame(columns=columns)
rows=[]
random.seed(10)
item_list=1000

for each in range(0,item_list):
    row=[]
    C_code=''.join(random.choices(string.ascii_uppercase + string.digits, k=random.randrange(6, 10)))
    
    MC1_df=random.choice(list(MC1))
    MC2_df=random.choice(MC1[MC1_df])
    Item_df=random.choice(MC2[MC2_df])
    row=[C_code,MC1_df,MC2_df,Item_df]
    
    for subcat_df in Item[Item_df][0]:
        row.append(random.choice(Subcat[subcat_df]))
    while len(row)<7:
        row.append("")
        
    unit=Item[Item_df][1]
    row.append(unit)
    
    rate=round(random.uniform(0.5, 40000), random.randrange(0, 4))
    row.append(rate)
    
    description=''.join(random.choices(string.ascii_uppercase + string.digits, k=random.randrange(6, 10)))+" "+''.join(random.choices(string.ascii_uppercase + string.digits, k=random.randrange(0, 3)))+" "+''.join(random.choices(string.ascii_uppercase + string.digits, k=random.randrange(0, 8)))
    row.append(description)

    rows.append(row)
```

```python
df=pd.DataFrame(rows)
df.columns=columns
df=df.fillna('')
df.sort_values(["Main Category 1","Main Category 2","Item","Sub Category 1","Sub Category 3"],ignore_index=True,inplace=True)
df.to_csv("Sample Extracted Library.csv",index=False)
```
